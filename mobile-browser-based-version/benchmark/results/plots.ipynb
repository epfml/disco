```python
%matplotlib inline
import matplotlib.pyplot as plt
import json
```

```python
def plot_measure(measure, data, xaxis = 'epochs'):
    
    for key in data:
        sample = data[key]
        plt.plot(sample[xaxis], sample[measure], label='user:{}'.format(key))
    
    plt.xlabel(xaxis)
    plt.ylabel(measure)
    plt.legend(loc="upper left")
```

```python
path = 'test.json'
```

```python
with open(path, 'r') as f:
  data = json.load(f)
```

```python
plot_measure('trainAccuracy', data)
```

```python
plot_measure('validationAccuracy', data)
```

```python
plot_measure('loss', data)
```
